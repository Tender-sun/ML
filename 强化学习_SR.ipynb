# 强化学习作业

让我们来模拟Q学习算法!假设存在状态0、1、2、3和动作('b'、'c')，并且折扣因子$\gamma =0.9$。此外，假设所有Q值初始化为0，并且学习率$\alpha =0.5,α=0.5$。

表格中的每一行t表示时间t的经验记录,$(s_t,a_t,r_t)$

在每行t，更新$Q(s_t,a_t)←q $将由$(s_t,a_t,r_t,s_{t+1})$基于Q学习算法完成。

Q-learning更新公式如下:
$Q(s,a)=(1-\partial )Q(s,a)+\partial(r+\gamma* \underset{a{}'}{max}Q(s{}',a{}'))$

```python
experience =  [(0, 'b', 0), #t = 0
              (2, 'b', 0),
              (3, 'b', 2),
              (0, 'b', 0), #t = 3
              (2, 'b', 0),
              (3, 'c', 2),
              (0, 'c', 0), #t = 6
              (1, 'b', 1),
              (0, 'b', 0),
              (2, 'c', 0), #t = 9
              (3, 'c', 2),
              (0, 'c', 0),
              (1, 'c', 1), #t = 12
              (0, 'c', 0),
              (2, 'b', 0),
              (3, 'b', 2), #t = 15
              (0, 'b', 0),
              (2, 'c', 0),
              (3, '', 0), #t = 18
              ]
```

示例程序：Q是每个状态和动作的矩阵关系，每个时刻都有一个Q矩阵，这个对应的关系就是每一个动作和状态的组合都在该时刻的矩阵中唯一对应一个值，下面我们把这个值卸载了后面问题描述中的第四列

```python
import numpy as np 
def lQ(exp,states,actions,alpha,gamma):
  Q = np.zeros((len(states),len(actions)))
  for i in range(len(exp)-1):
    k = -1
    if exp[i][1] == 'b':
      k=0
    else :
      k=1
    Q[exp[i][0],k] = (1-alpha)*Q[exp[i][0],k]+alpha*(exp[i][2]+gamma*np.max(Q[exp[i+1][0],:]))
    print("t=",i,"\n","Q=",Q.T)
  return Q
lQ(experience,[0,1,2,3],['b','c'],0.5,0.9)
```
